```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# read all data from csv
df = pd.read_csv('survey_results_public.csv')

df.head()
```

```python
# read 3 column data which is age, DevType and RemoteWork from 44 row to 55
data = df.loc[44:55, ['Age', 'DevType', 'RemoteWork']]

data
```

```python
# set responseid as an index for table
indexed_table = df.set_index('ResponseId')

indexed_table
```

```python
# filter data which is filtered by remotework and devtype
data_filtered = df[(df['RemoteWork'] == 'Remote') & (df['DevType'] == 'Developer, back-end')]

data_filtered
```

```python
# sort data with convertedCompYearly and age
table_sorted = df.sort_values(by=['ConvertedCompYearly', 'Age'])

table_sorted
```

```python
# show mean, standart, median, min and max for responseId
response_id_data = df['ResponseId'].dropna()
mean = response_id_data.mean()
standart = response_id_data.std()
median = response_id_data.median()
min = response_id_data.min()
max = response_id_data.max()

mean, standart, median, min, max

```

```python
salary = df.groupby('DevType')['ConvertedCompYearly'].mean().dropna()
salary
```

```python
# show statistic using bar, for devtype in x label and average salary in usd in y label
plt.figure(figsize=(10, 6))
plt.bar(salary.index, salary.values, color='green')
plt.xlabel('Devtype Group')
plt.ylabel('Average Salary USD')
plt.title('Average Salary by Devtype Group')
plt.xticks(rotation=90)
plt.show()
```

```python
# 
age_group = df.groupby('Age')['ConvertedCompYearly'].mean().dropna()

age_group
```

```python
# show data with line plot. x label is age group and y label is avearge salary in usd
plt.figure(figsize=(12, 6))
plt.plot(age_group.index, age_group.values, marker='o', linestyle='-', color='green')
plt.xlabel('Age Group')
plt.ylabel('Average Salary USD')
plt.title('Average Salary by Age Group')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
